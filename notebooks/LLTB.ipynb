```python
import numpy as np
import pandas as pd
import subprocess
from subprocess import Popen,PIPE
import matplotlib
import matplotlib.pyplot as plt
import os
import warnings
from classy import Class
from scipy.interpolate import interp1d
import scipy.integrate as integrate
import scipy.constants as const
from LLTB_functions import metric_functions, derivate_functions
```

# $\Lambda$CDM - Background cosmology

```python
omega_b = 2.2253e-02
omega_cdm =0.11919
h =  0.6835
Omega_k = 0.0
N_ur = 3.046
ln10A_s = 3.07
n_s = 0.9619
```

```python
cosmo = Class()
cosmo.set({'lensing': 'yes', 'output': 'tCl lCl pCl', 'h': h, 'omega_cdm': omega_cdm, 'Omega_k': Omega_k,
           'omega_b': omega_b, 'N_ur': 3.046, 'ln10^{10}A_s': ln10A_s, 'n_s': n_s, 'thermodynamics_verbose':1})
cosmo.compute()

zdrag = cosmo.get_current_derived_parameters(['z_d'])['z_d']
YHe = cosmo.get_current_derived_parameters(['YHe'])['YHe']
zreio = cosmo.get_current_derived_parameters(['z_reio'])['z_reio']
```

```python
cc = const.physical_constants['speed of light in vacuum'][0]/1000.
dAz_LCDM = np.vectorize(cosmo.angular_distance)
Hz_LCDM = np.vectorize(cosmo.Hubble)

def dLz_LCDM(z):
    dl = (1.0+z)**2*dAz_LCDM(z)
    return dl

def muz_LCDM(z):
    mu = 5*np.log10(dLz_LCDM(z))+25
    return mu

def DeltaThetaz_LCDM(z):
    deltaThe = cosmo.rs_drag()/((1+z)*dAz_LCDM(z))
    return deltaThe

def DeltaZz_LCDM(z):
    deltaz = cosmo.rs_drag()*cc*Hz_LCDM(z)
    return deltaz
```

# $\Lambda$LTB cosmology

```python
path_vd2020 = '/path-to-vd2020'

H0 = 100*h
Omega_DE = cosmo.Omega_Lambda()
delta0 = 0.75
zB = 0.5

inho_param = (H0,omega_b,omega_cdm,Omega_k,Omega_DE,delta0,zB,zdrag,YHe,zreio)
cmd = path_vd2020+'/wLLTBBackground/bin/vd2020 c %.4f %.10f %.8f %.8f %.6f -1.0 %.8f %.6f %.6f %.6f %.6f'%inho_param
```

```python
nonsense_cosmo = 0

process = subprocess.Popen([cmd],shell=True,stdout=PIPE,stdin=PIPE)
# getting whole output
outvd2020_ini=process.communicate()[0].split('\n') # This is already a list where each element is a line output
process.wait()
# debbuging point with numerical issues
try:
    int(outvd2020_ini[-2])
except:
    nonsense_cosmo += 1
    warnings.warn('Sampled point crash vd2020:')
    warnings.warn(cmd)
    warnings.warn('It is assumed -loglkl = 1e5 by experiment (for Planck 1e8)')

if nonsense_cosmo == 0:
    # rL and rBMpc
    for i, line in enumerate(outvd2020_ini):
        if 'r/L at which delta density changes sign' in line:
                line = line.split()
                LoverB = float(line[-1])
        if 'Furthest border:' in line:
                line = line.split()
                rBMpc = float(line[-1])

    # Exception when zB or delta0 are equal to zero
    try:      
        rBMpc 
    except:
        rBMpc = 0.
        LoverB = 0.

    # len z
    lenz=int(outvd2020_ini[-2]) #-2 because the last lines it's a empty line
    lennuinsancelines=3
    # removing the first lines/comments...
    outvd2020=outvd2020_ini[-lenz-lennuinsancelines:][:-lennuinsancelines+1]
    # split lines string to get a list
    outvd2020=[x.split() for x in outvd2020]
    # re-arrange to get a dataframe
    # outvd2020[0] contains the name of the distances and metric functions
    outvd2020_col=pd.DataFrame(np.array(outvd2020[1:],dtype=float),index=np.arange(len(outvd2020)-1),columns=outvd2020[0])

    # looking where effective parameters are
    eff_pos = np.where(np.array(outvd2020_ini) == ' EffectiveParams:')[0][0]
    # getting the effective parameters itself
    eff_params = outvd2020_ini[eff_pos:eff_pos+3]
    eff_params = [x.split() for x in eff_params]
    outvd2020_params = pd.DataFrame(np.array(eff_params[2:],dtype=float),columns=eff_params[1])
    # including rL/rB and rBMpc
    outvd2020_params['LoverB'] = LoverB
    outvd2020_params['rBMpc'] = rBMpc
else:
    outvd2020_col = np.array([])
    outvd2020_params = np.array([])
```

### Effective $\Lambda$CDM cosmology

```python
h_eff = outvd2020_params['H0_eff'][0]/100.0
omega_b_eff = outvd2020_params['omegah2_b_eff'][0]
omega_dm_eff = outvd2020_params['omegah2_dm_eff'][0]
OmegaK_eff = outvd2020_params['OmegaK_eff'][0]
T0_eff = outvd2020_params['Tcmb_eff'][0]

newcosmo_arguments = {'lensing': 'yes', 'output': 'tCl lCl pCl','h':h_eff, 'Omega_k': OmegaK_eff,
                      'omega_cdm': omega_dm_eff, 'omega_b': omega_b_eff, 'T_cmb': T0_eff,'N_ur': 3.046, 
                      'ln10^{10}A_s': 3.07, 'n_s': 0.9619}
newcosmo = Class()
newcosmo.set(newcosmo_arguments)

newcosmo.compute()
rs_eff = newcosmo.rs_drag()

zdrag_eff = newcosmo.get_current_derived_parameters(['z_d'])['z_d']
```

### Metric functions, distances, observables and $\Omega$s

```python
LLTB_dic = {}
LLTB_dic['zB'] = zB
LLTB_dic['VH0'] = H0
LLTB_dic['VOmegaDE'] = Omega_DE
LLTB_dic['VOmegaK'] = Omega_k
LLTB_dic['MB'] = -19.24
LLTB_dic['rs_eff'] = rs_eff
LLTB_dic['zdrag_eff'] = zdrag_eff
```

```python
mfun = metric_functions(outvd2020_col)
dfun = derivate_functions(outvd2020_col,LLTB_dic,bao=True)
```
